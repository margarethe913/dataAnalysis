```python
from collections import defaultdict

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('responses.csv')
#make sure user answered all eligibility questions
ageGroup = df.columns.get_loc("Q155")
gender = df.columns.get_loc("Q145")
country = df.columns.get_loc("Q151")

school = df.columns.get_loc("Q153")
professional = df.columns.get_loc("Q157")
major = df.columns.get_loc("Q159")
role = df.columns.get_loc("Q156")
experience = df.columns.get_loc("Q6")
lastApp = df.columns.get_loc("Q8")
download = df.columns.get_loc("Q158")
appTeamSize = df.columns.get_loc("Q163")
privacyKnowledge = df.columns.get_loc("Q160")
participants = []
ages = []
genders = []
countries = []
schools = []
professionals = []
majors = []
roles = []
experiences = []
lastApps = []
downloads = []
appTeamSizes = []
privacyKnowledges =[]
for j in range(2, len(df.iloc[:, 0])):
    participants.append(df.iat[j, df.columns.get_loc("Q162")])
    ages.append(df.iat[j, ageGroup])
    genders.append(df.iat[j, gender])
    countries.append(df.iat[j, country])
    schools.append(df.iat[j, school])
    professionals.append(df.iat[j, professional])
    majors.append(df.iat[j, major])
    roles.append(df.iat[j, role])
    experiences.append(df.iat[j, experience])
    lastApps.append(df.iat[j, lastApp])
    downloads.append(df.iat[j, download])
    appTeamSizes.append(df.iat[j, appTeamSize])
    privacyKnowledges.append(df.iat[j, privacyKnowledge])
```

```python
results = pd.DataFrame()
results['Participant ID'] = participants
results['Age Group'] = ages
results['Gender'] = genders
results['Country'] = countries
results['Education'] = schools
results['Professional Experience'] = professionals
results['College Major'] = majors
results['Development Role'] = roles
results['Years of Experience'] = experiences
results['Last Published App'] = lastApps
results['App downloads'] = downloads
results['App Team Size'] = appTeamSizes
results['Privacy Label Familiarity'] = privacyKnowledges
results.to_csv("demographics.csv")
```
